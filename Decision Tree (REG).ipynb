Decision tree regression 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor , plot_tree
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv("Position_Salaries.csv")
```

```python
data.head(5)
```

```python
le = LabelEncoder()
data["Position"] = le.fit_transform(data["Position"])
```

```python
data.head(3)
```

```python
sns.pairplot( data = data)
plt.show
```

```python
x = data.iloc[:,:-1]
y = data["Salary"]
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.25 , random_state = 42)
```

```python
dtr = DecisionTreeRegressor(max_depth = 5) #pre pruning
dtr.fit(x_train , y_train)
```

```python
plt.figure(figsize = (5 , 4))
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf = dtr)
plt.show()
```

```python
dtr.score(x_train , y_train)*100 , dtr.score(x_test , y_test)*100 #bad accuracy bcz no proper data is available for dtr
```

```python
plt.figure(figsize = (10 , 9))
plot_tree(dtr , filled = True)
plt.show
```
