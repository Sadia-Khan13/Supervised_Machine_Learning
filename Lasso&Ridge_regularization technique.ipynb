```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.linear_model import Lasso , Ridge
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
columns = ["Suburb" , "Address" , "Rooms" , "Type" , "Price" , "Method"  , "SellerG" , "Distance" , "Bedroom2" , "Bathroom"  , 
          "Landsize" , "BuildingArea" , "YearBuilt" , "CouncilArea" ,"Lattitude" , "Longtitude" , "Regionname"]  
data = data[columns]
data.head()
```

```python
data.isnull().sum()
```

```python
columns_null = ["Bedroom2" , "Bathroom" ]
data[columns_null] = data[columns_null].fillna(0)
```

```python
null = ["Landsize" , "BuildingArea" , "YearBuilt" , "Lattitude" , "Longtitude"]
data[null] = data[null].fillna(data[null].mean())
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
le = LabelEncoder()
for i in data.select_dtypes(include =object).columns:
    data[i] = le.fit_transform(data[i])
```

```python
data.head()
```

```python
x = data.drop("Price", axis = 1)
y = data["Price"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 50)
```

```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train)*100
#lr.score(x_test , y_test)
```

```python
lr.score(x_test , y_test)*100
```

In linear regression data is underfit

```python
ll = Lasso(alpha = 40)
ll.fit(x_train , y_train)
```

```python
ll.score(x_train , y_train)
```

```python
ll.score(x_test , y_test)
```

```python
RR = Ridge()
RR.fit(x_train , y_train)
```

```python
RR.score(x_train , y_train)
```

```python
RR.score(x_test , y_test)
```
