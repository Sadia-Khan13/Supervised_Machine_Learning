# Logistic Regression (Multi class classification)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

le = LabelEncoder()
le.fit(data["species"])
data["species"]= le.transform(data["species"])

```python
data.head()
```

```python
x = data.iloc[: , :-1]
y = data["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,  y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

# Multi class using "ovr" method

```python
lr = LogisticRegression(multi_class = "ovr")
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train)*100
```

```python
lr.score(x_test , y_test)*100
```

```python
lr.predict(x_test)
```

```python
lr.predict([[5.1 , 3.5	,1.4	,0.2]])
```

```python
lr.predict([[6.7 ,3.0 ,	5.2 ,2.3]])
```

# Using Multinomial method

```python
lrr = LogisticRegression(multi_class = "multinomial")
lrr.fit(x_train , y_train)
```

```python
lrr.score(x_train , y_train)*100
```

```python
lrr.score(x_test , y_test)*100
```

```python
lrr.predict(x_test)
```

```python
lrr.predict([[5.1 , 3.5	,1.4	,0.2]])
```

```python
lrr.predict([[6.7 ,3.0 ,5.2 ,2.3]])
```

Input features was not valid but we just try to understand how logistic reg work with multiple features
