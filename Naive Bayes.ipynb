```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv("job placement.csv")
```

```python
data.head()
```

```python
data.drop(columns =["Student_ID"] ,axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data["IQ"]) #Checking the input features are normal or not
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot(x="CGPA" ,y="IQ" , data=data , hue = "Placement")
plt.show()
```

```python
x = data.iloc[: , :-1]
y = data["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_train , y_train)*100 , gnb.score(x_test , y_test)*100
```

```python
plt.figure(figsize = (4,3))
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train , y_train)
```

```python
mnb.score(x_train , y_train)*100 , mnb.score(x_test , y_test)*100
```

```python
plt.figure(figsize = (4,3))
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(x_train , y_train)
```

```python
bnb.score(x_train , y_train)*100 , bnb.score(x_test , y_test)*100
```

```python
plt.figure(figsize = (4,3))
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=bnb)
plt.show()
```

```python
gnb.predict([[5.8 , 50]])
```
