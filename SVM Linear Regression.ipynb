```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
data = pd.read_csv("Placement.csv")
data.head()
```

```python
#checking the data is linear or non linear
plt.figure(figsize=(5 , 4))
sns.scatterplot(x="cgpa" , y="package" , data=data)
plt.show()
```

```python
x = data[["cgpa"]] #data must be 2 dimentional
y = data["package"]
```

```python
#spliting data into training and testing
x_train , x_test , y_train , y_test = train_test_split(x , y  , test_size=0.2 , random_state=26)
```

```python
svr = SVR(kernel="rbf")
svr.fit(x_train , y_train)
```

```python
svr.score(x_train , y_train)*100 , svr.score(x_test , y_test)*100
```

```python
plt.figure(figsize=(5 , 4))
sns.scatterplot(x="cgpa" , y="package" , data=data)
plt.plot(data["cgpa"] , svr.predict(x) , c="red" )
plt.legend(["org data" , "prediction line"])
plt.show()
```
